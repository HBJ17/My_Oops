```python
#with init 
#__init__ is an initializer for a class instance (object). It is automatically called every time a new object is created from a class. 

class F1:
    def __init__(self, team, driver):
        self.team = team
        self.driver = driver

    def show(self):
        print(f"Team = {self.team}")
        print(f"Driver = {self.driver}")

t1 = F1("Red Bull", "Max Verstappen")
t2 = F1("Ferrari", "Lewis Hamilton")

t1.show()
print()
t2.show()
```

```python
#mini project
#Banking System

class Bankaccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance = self.balance + amount
            print(f"Deposit of {amount} made.")
        else:
            print("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > 0:
            self.balance = self.balance - amount
            print(f"Withdrawal of {amount} made.")
        else:
            print("Withdrawal amount must be positive")
        
    def show_balance(self):
        print(f"Current balance: {self.balance}")

    def show_owner(self):
        print(f"Owner: {self.owner}")

b1 = Bankaccount("Sakthi",1000)
b1.deposit(1000)
b1.withdraw(500)
b1.show_balance()
b1.show_owner()
```
