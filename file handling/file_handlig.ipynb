```python
f = open("data.txt", "r")
content = f.read()
print(content)
f.close()
```

```python
f = open("data.txt", "r")
lines = f.readlines()
print(lines)
f.close()
```

```python
f = open("data.txt", "r")
for line in f:
    print(line)
```

```python
f = open("write.txt", "w")
f.write("Hello Python\n")
f.write("File handling example")
f.close()

f = open("write.txt", "r")
print(f.read())
f.close()
```

```python
f = open("write.txt", "a")
f.write("\nThis line is appended")
f.close()

f = open("write.txt", "r")
print(f.read())
f.close()
```

```python
f = open("newfile.txt", "x")
f.write("New file created")
f.close()
```

```python
with open("file.txt", "r") as f:
    content = f.read()
    print(content)
```

```python
f = open("data.txt")
print(f.name)    # File name
print(f.mode)    # Mode
print(f.closed)  # True / False
f.close()
```
